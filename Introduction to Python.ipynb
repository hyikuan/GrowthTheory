## Exercise 1

```python
def factorial(n):
    s = 1
    for i in range(n):
        s = s* (i+1)
    return s
```

```python
factorial(5)
```

## Exercise 2

```python
from numpy.random import uniform
def binomial_rv(n, p):
    re = 0
    for i in range(n):
        x = uniform()
        if x<p:
            re += 1
    return re
```

```python
binomial_rv(1000,0.5)
```

## Exercise 3

```python
import numpy as np
count = 0
n = 100
for i in range(n):
    x = np.random.uniform()
    y = np.random.uniform()
    if (x*x+y*y)<1:
        count += 1
print((count*4)/n)
```

## Exercise 4

```python
from numpy.random import uniform
def flip_coins():
    count = 0
    for i in range(10):
        x = uniform()
        if x<0.5:
            count += 1
        else:
            count = 0
        if count>=3: break
    if count >= 3: return('pay')
    else: return('nothing')
```

```python
flip_coins()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
def TS(T, alpha):
    series = []
    series.append(0)
    for i in range(T):
        epsilon = np.random.randn()
        series.append(alpha*series[i]+epsilon)
    return series
S1 = TS(T, alpha)
plt.plot(S1)
```

## Exercise 6

```python
for alpha in [0, 0.8, 0.98]:
    S = TS(200, alpha)
    plt.plot(S, label = 'alpha = '+ str(alpha))
plt.legend()
plt.show()
```

## Exercise 4

```python
def f(seq_a, seq_b):
    is_subset = True
    for a in seq_a:
        if a not in seq_b:
            is_subset = False
    return is_subset
```

```python
def judge(a, b):
    key1 = True
    for i in a:
        key2 = False
        for j in b:
            if i == j:
                key2 = True 
        if not key2:
            key1 = False
    return key1
```

```python
a = 'acbeccd'
b = 'abcd'
judge(a, b)
```

## Exercise 5

```python
def linapprox(f, a, b, n, x):
    step = (b-a)/(n-1)
    u, v = int((x-a)/step)*step+a, (int((x-a)/step)+1)*step+a
    return f(u) + (x - u) * (f(v) - f(u)) / (v - u)
```

```python
def test(f, a, b, n, x):
    length_of_interval = b - a
    num_subintervals = n - 1
    step = length_of_interval / num_subintervals
    point = a
    while point <= x:
        point += step
    u, v = point - step, point
    return f(u) + (x - u) * (f(v) - f(u)) / (v - u)
```

```python
a = 0
b = 10
f = lambda x: x**3
n = 10
```

```python
test(f, a, b, n, 4.5)
```

```python
linapprox(f, a, b, n, 4.5)
```
